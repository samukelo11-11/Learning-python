```python
!pip install transformers accelerate


from transformers import pipeline

# This downloads a lightweight, smart AI model from Microsoft called Phi-3
ai_model = pipeline("text-generation", model="microsoft/Phi-3-mini-4k-instruct", device_map="auto")

# Give it a prompt
prompt = "Write a short, motivational quote for a 16-year-old programmer."

# Run the model
messages = [{"role": "user", "content": prompt}]
output = ai_model(messages, max_new_tokens=100)

print(output[0]['generated_text'])

```
